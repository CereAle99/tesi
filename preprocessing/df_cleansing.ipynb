```python
import os
import csv
import pandas as pd
```

MM PET features dataset cleansing

```python
total_index = pd.read_csv("../data/tot_MM_patientes.csv", sep=',')

tot_feat_index = pd.read_csv("../data/raw_MM_PT_feats.csv", sep=',')
```

```python
tot_feat_index = tot_feat_index.drop("Unnamed: 0", axis=1)
```

```python
original_feats = tot_feat_index[tot_feat_index["mask_shape"] == "original"]
fill_holes_feats = tot_feat_index[tot_feat_index["mask_shape"] == "fill_holes"]
dilation_feats = tot_feat_index[tot_feat_index["mask_shape"] == "dilation"]
```

```python
original_feats
```

```python
original_feats = original_feats.sort_values('patient_id')
original_feats = original_feats.drop_duplicates('patient_id', keep='last')
fill_holes_feats = fill_holes_feats.sort_values('patient_id')
fill_holes_feats = fill_holes_feats.drop_duplicates('patient_id', keep='last')
dilation_feats = dilation_feats.sort_values('patient_id')
dilation_feats = dilation_feats.drop_duplicates('patient_id', keep='last')
```

```python
original_feats.to_csv("../data/MM_PET_feats_orig.csv", index=False, index_label="patient_id")
fill_holes_feats.to_csv("../data/MM_PT_feats_fill.csv", index=False, index_label="patient_id")
dilation_feats.to_csv("../data/MM_PT_feats_dil.csv", index=False, index_label="patient_id")
```

```python
total_feats_df = pd.concat([original_feats, fill_holes_feats, dilation_feats], axis=0)
total_feats_df.to_csv("../data/MM_PT_feats.csv", index=False)
```

Healty PET feature dataset cleansing

```python
raw_data = pd.read_csv("../data/raw_healty_PT_feats.csv", sep=',')

ambiguous = pd.read_csv("../log/healthy_ambiguous_segm.csv", sep=',')
# ambiguous = ambiguous.T
```

```python
raw_data.head()
```

```python
ambiguous.head()
```

```python
original_feats = raw_data[raw_data["mask_shape"] == "original"]
fill_holes_feats = raw_data[raw_data["mask_shape"] == "fill_holes"]
dilation_feats = raw_data[raw_data["mask_shape"] == "dilation"]
```

```python
original_feats
```

```python
original_feats = original_feats.sort_values('patient_id')
original_feats = original_feats.drop_duplicates('patient_id', keep='last')
fill_holes_feats = fill_holes_feats.sort_values('patient_id')
fill_holes_feats = fill_holes_feats.drop_duplicates('patient_id', keep='last')
dilation_feats = dilation_feats.sort_values('patient_id')
dilation_feats = dilation_feats.drop_duplicates('patient_id', keep='last')
```

```python
index_orig = original_feats.set_index('patient_id').index
ambiguous = [idx for idx in ambiguous if idx in index_orig]
print(len(ambiguous))
original_feats = original_feats[~original_feats['patient_id'].isin(ambiguous)]
```

```python
original_feats.to_csv("../data/healthy_PT_feats_orig.csv", index=False, index_label="patient_id")
fill_holes_feats.to_csv("../data/healthy_PT_feats_fill.csv", index=False, index_label="patient_id")
dilation_feats.to_csv("../data/healthy_PT_feats_dil.csv", index=False, index_label="patient_id")
```

```python
total_feats_df = pd.concat([original_feats, fill_holes_feats, dilation_feats], axis=0)
total_feats_df.to_csv("../data/healthy_PT_feats.csv", index=False)
```
